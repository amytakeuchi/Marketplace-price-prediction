```python
train = pd.read_csv('/Users/asamitakeuchi/mercari/train2.csv')
```

## Exploring the text data

```python
!install wordcloud
from wordcloud import WordCloud, STOPWORDS
text = train['item_description'][:100000].str.cat(sep=' ')
wordcloud = WordCloud().generate(text)
# Display the generated image:
plt.imshow(wordcloud, interpolation='None')
plt.axis("off")
plt.show()
```

Let's see the frequently appearing words by diffrent price range.

```python
#price range will be examined lower range vs upper range
#lower range: 0 - 100, 101 - 200, 201 - 500
#upper range: 500 - 1,000, 1,001 - 1,500, 1,501<
low_0_100 = train[(train['price'] >= 0) & (train['price'] <= 100)]
low_101_200 = train[(train['price'] >= 101) & (train['price'] <= 200)]
low_201_500 = train[(train['price'] >= 201) & (train['price'] <= 500)]

text_0_100 = low_0_100['item_description'].str.cat(sep=' ')
text_101_200 = low_101_200['item_description'].str.cat(sep=' ')
text_201_500 = low_201_500['item_description'].str.cat(sep=' ')

wordcloud = WordCloud().generate(text_0_100)
wordcloud2 = WordCloud().generate(text_101_200)
wordcloud3 = WordCloud().generate(text_201_500)

plt.imshow(wordcloud, interpolation='None')
plt.title('Frequent words from price range between $0 - 100')
plt.axis('off')
plt.show()
```

```python
plt.imshow(wordcloud2, interpolation='None')
plt.title('Frequent words from price range between $101 - 200')
plt.axis('off')
plt.show()
```

```python
plt.imshow(wordcloud3, interpolation='None')
plt.title('Frequent words from price range between $200 - 500')
plt.axis('off')
plt.show()
```

```python
#word clouds for upper range prices
up_500_1000 = train[(train['price'] >= 500) & (train['price'] <= 1000)]
up_1001_1500 = train[(train['price'] >= 1001) & (train['price'] <= 1500)]
up_1501 = train[(train['price'] >= 1501)]

text_500_1000 = up_500_1000['item_description'].str.cat(sep=' ')
text_1001_1500 = up_1001_1500['item_description'].str.cat(sep=' ')
text_1501 = up_1501['item_description'].str.cat(sep=' ')

wordcloud4 = WordCloud().generate(text_500_1000)
wordcloud5 = WordCloud().generate(text_1001_1500)
wordcloud6 = WordCloud().generate(text_1501)

plt.imshow(wordcloud3, interpolation='None')
plt.title('Frequent words from price range between $500 - 1,000')
plt.axis('off')
plt.show()
```

```python
plt.imshow(wordcloud5, interpolation='None')
plt.title('Frequent words from price range between $1,001 - 1,500')
plt.axis('off')
plt.show()
```

```python
plt.imshow(wordcloud6, interpolation='None')
plt.title('Frequent words from price range above $1,501')
plt.axis('off')
plt.show()
```
