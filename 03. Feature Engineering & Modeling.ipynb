```python
#check null data
def check_if_null(df):
    null = df.isnull().sum()
    null_percent = (df.isnull().sum()/df.count()*100).round(2)
    null_values = pd.concat([null, null_percent], axis = 1)
    null_values.columns = ['Null Values', '%Null']
    return null_values

check_if_null(train)
```

## Encoding by column transformer
I am going to use ColumnTransformer to apply LabelBinarizer, CountVectorizer, TfidfVectorizer, and Normaliaer to encode categorical and numerical variables.

```python
#drop target variable and create y
X = train_prep.drop('price', axis=1)  # Input features
y = train_prep['price']  # Target variable

X.isnull().sum()
```

```python
y.isnull().sum()
```

```python
#Split train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=.2, random_state=42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
num_vals = ['name_word_count', 'desc_word_count', 'desc_word_len']
#text_vals = ['name','brand_name','cat_2_3']
#text_desc = ['item_description']

scaler = StandardScaler()

cv = CountVectorizer(
    ngram_range=(1, 2), 
    max_features=100000, 
    token_pattern = r'(?u)\w+|[^\w\s]',
)

tv = TfidfVectorizer(max_features=50000,
                     ngram_range=(1, 2),
                     stop_words='english')

col_transformer = ColumnTransformer(
    transformers=[
      ('num_vals',scaler, num_vals),
      ('name_vals', cv, 'name'),
      ('brand_name_vals', cv, 'brand_name'),
      ('cat_2_3_vals', cv, 'cat_2_3'),
      ('text_desc', tv, 'item_description')])

vectorizer = col_transformer.fit(X_train)
train_vec = vectorizer.transform(X_train).astype(np.float32)
test_vec = vectorizer.transform(X_test).astype(np.float32)

print(train_vec.shape)
print(test_vec.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
print(X.shape)  
print(len(y)) 
```

##  Ridge Regression

```python
#fit the data
#Ridge regression
ridge_reg = Ridge(solver = "lsqr", fit_intercept=True)

ridge_reg.fit(train_vec, y_train)
```

```python
#train and test scores for Ridge regression
train_score_ridge = ridge_reg.score(train_vec, y_train)
test_score_ridge = ridge_reg.score(test_vec, y_test)
cv_scores_ridge = cross_val_score(
    ridge_reg, train_vec, y_train, cv=5,
    scoring='neg_root_mean_squared_error'  # Use negative RMSE as the scoring metric
)
cv_rmse_scores_ridge = -cv_scores_ridge

print("\nRidge Model............................................\n")
print("The train score for ridge model is {}".format(train_score_ridge))
print("The test score for ridge model is {}".format(test_score_ridge))
print("Cross-Validation RMSE Scores:", cv_rmse_scores_ridge)
print("Mean Cross-Validation RMSE:", cv_rmse_scores_ridge.mean())
```

```python
# Predict on the test set
y_pred_ridge = ridge_reg.predict(test_vec)

# Calculate residuals
#Residuals of Ridge 
residuals_ridge = y_test - y_pred_ridge

# Plot Residuals vs Predicted Values
fig, ax1 = plt.subplots(figsize=(16, 6))

ax1.scatter(y_pred_ridge, residuals_ridge, alpha=0.6)
ax1.set_title("Residuals vs Predicted Values of Ridge Regression")
ax1.set_xlabel("Predicted Values")
ax1.set_ylabel("Residuals")
ax1.axhline(y=0, color='r', linestyle='--')

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```

```python
# Plot Q-Q Plot
fig, ax1 = plt.subplots(figsize=(16, 6))
#Ridge
ax1 = stats.probplot(residuals_ridge, plot=plt)
ax1 = plt.title("Q-Q Plot of Residuals of Ridge Regression")
ax1 = plt.xlabel("Theoretical Quantiles")
ax1 = plt.ylabel("Sample Quantiles")
ax1 = plt.show()
```
